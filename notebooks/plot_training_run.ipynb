```python
# Example line of the training log file
"""Epoch    0 Loss -238.61 nats, Val. loss -636.07 nats.  MSE 0.4974. SSE 1.4922.  Val. MSE 0.2796. Val. SSE 0.8387."""

# open the log file
with open('epochs_stats.log', 'r') as f:
    lines = f.readlines()
    
# extract the following values (epoch, loss, val_loss, mse, sse, val_mse, val_sse)
epochs = []
losses = []
val_losses = []
mses = []
sses = []
val_mses = []
val_sses = []
for line in lines:
    epochs.append(int(line.split()[1]))
    losses.append(float(line.split()[3]))
    val_losses.append(float(line.split()[7]))
    mses.append(float(line.split()[10][:-1]))
    sses.append(float(line.split()[12][:-1]))
    val_mses.append(float(line.split()[15][:-1]))
    val_sses.append(float(line.split()[18][:-1]))
    
# plot the training and validation losses
import matplotlib.pyplot as plt
plt.plot(epochs, losses, label='Training loss')
plt.plot(epochs, val_losses, label='Validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# plot the training and validation MSE
plt.plot(epochs, mses, label='Training MSE')
plt.plot(epochs, val_mses, label='Validation MSE')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.legend()
plt.show()

# plot the training and validation SSE
plt.plot(epochs, sses, label='Training SSE')
plt.plot(epochs, val_sses, label='Validation SSE')
plt.xlabel('Epoch')
plt.ylabel('SSE')
plt.legend()
plt.show()

    

    
```
