```python
# import pickled line data
import pickle
with open('../data/data/chumba_v2/lines/3.pkl', 'rb') as f:
    line_data = pickle.load(f)
strokes = line_data['strokes']
```

```python
from data.utils import plot_line_from_strokes

# plot strokes
plot_line_from_strokes(strokes)
```

```python
# Take only the first 10 strokes
strokes = strokes[:12]
```

```python
# plot strokes
plot_line_from_strokes(strokes)
```

```python
# Flatten the strokes (turning all points into (x, y, eos) tuples)
# Last point of each stroke is (0, 0, 1)
points = []
print(len(strokes))
for stroke in strokes:
    for i, point in enumerate(stroke):
        if i == len(stroke) - 1:
            points.append((point[0], point[1], 1))
        else:
            points.append((point[0], point[1], 0))
            
print(points)
```

```python
# turn points to offsets
def to_offsets(points):
    if not points:
        return []

    offsets = []

    first_point = points[0]
    prev = first_point[0], first_point[1]

    for x, y, eos in points:
        prev_x, prev_y = prev
        offsets.append((x - prev_x, y - prev_y, eos))
        prev = x, y

    return offsets
offsets = to_offsets(points)
print(offsets)
```
